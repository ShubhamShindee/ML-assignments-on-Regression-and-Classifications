## Implementation of Polynomial Regression

```python
x=df.drop(columns="bodyfat")
y=df["bodyfat"]
```

```python
## Data Splitting into train and test
from sklearn.model_selection import train_test_split 
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=0)
from sklearn.linear_model import LinearRegression
```

```python
## Model Training
from sklearn.preprocessing import PolynomialFeatures
pf= PolynomialFeatures(degree=2)
x_pf= pf.fit_transform(xtrain,xtest)
model=LinearRegression()
model.fit(x_pf, ytrain)
```

```python
## Model Testing
ypred =model.predict(pf.fit_transform(xtest))
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
## Model Evaluation
mse=mean_squared_error(ytest,ypred) 
print('mse:',mse)
mae=mean_absolute_error(ytest,ypred)
print("mae:",mae)
rmse=np.sqrt(mse)
print("rmse:",rmse)
r2=r2_score(ytest,ypred)
print("r2_score:",r2)
```

```python
## Model Training on degree 3
pf= PolynomialFeatures(degree=3)
x_pf= pf.fit_transform(xtrain,xtest)
model=LinearRegression()
model.fit(x_pf, ytrain)
```

```python
ypred =model.predict(pf.fit_transform(xtest))
```

```python
mse=mean_squared_error(ytest,ypred) 
print('mse:',mse)
mae=mean_absolute_error(ytest,ypred)
print("mae:",mae)
rmse=np.sqrt(mse)
print("rmse:",rmse)
r2=r2_score(ytest,ypred)
print("r2_score:",r2)
```

#### r square value is near to 1 as we decrease the degree for outliers contain dataset.From the above comparison it seems that if we increase the degree of polynomial regression the r square value get closer to 0 which is not good for our model.

### PR on Outliers removed data

```python
## Removing outliers
## IQR== difference between 3rd quartile and 1st quartile >=Q3-Q1
Q1=df.quantile(0.25)
Q3=df.quantile(0.75)
IQR=Q3-Q1
IQR
```

```python
# Q1-1.5*IQR to Q3+1.5*IQR
df1=df[~((df<(Q1-1.5*IQR)) | (df> (Q3+1.5*IQR))).any(axis=1)]
df1.shape
```

```python
df1.boxplot(figsize=(12,6))
```
